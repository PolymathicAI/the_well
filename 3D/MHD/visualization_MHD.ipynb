```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
bp = ['.1', '1']
pp = ['.01', '.1', '.2', '1', '2']
b = bp[1]
p = pp[0]
path = f'data/b{b}_p{p}.hdf5'
```

```python
#data is (10, 100, 256, 256, 256)
with h5py.File(path, 'r') as f:
    density = f['density'][0]
    mag_x = f['mag_x'][0]
    mag_y = f['mag_y'][0]
    mag_z = f['mag_z'][0]
    vel_x = f['velocity_x'][0]
    vel_y = f['velocity_y'][0]
    vel_z = f['velocity_z'][0]
```

```python
slice_nb = 128
time = 50
fig, axs = plt.subplots(3, 3, figsize=(15, 15))
axs[0, 0].imshow(density[time, slice_nb], cmap='viridis')
axs[0, 0].set_title('Density')
axs[0, 1].imshow(mag_x[time, slice_nb], cmap='viridis')
axs[0, 1].set_title('Magnetic field x')
axs[0, 2].imshow(vel_x[time, slice_nb], cmap='viridis')
axs[0, 2].set_title('Velocity x')
axs[1, 0].imshow(mag_y[time, slice_nb], cmap='viridis')
axs[1, 0].set_title('Magnetic field y')
axs[1, 1].imshow(mag_z[time, slice_nb], cmap='viridis')
axs[1, 1].set_title('Magnetic field z')
axs[1, 2].imshow(vel_y[time, slice_nb], cmap='viridis')
axs[1, 2].set_title('Velocity y')
axs[2, 0].imshow(vel_z[time, slice_nb], cmap='viridis')
axs[2, 0].set_title('Velocity z')
plt.show()
```
