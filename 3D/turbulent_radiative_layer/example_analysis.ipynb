```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import h5py as h5
```

for a given simulation there are 5 state variables rho, P, vel1, vel2, vel3

there are 9 different cooling time parameters ranging from 0.03 to 3.16 (log spaced)

for each cooling time there are 10 different initial random seeds for the noise

the data is arranged as (n_seeds, n_times, 256, 128, 128)

the spatial axes are labeled 3, 2, 1 and correspond to the 256, 128, and 128 indexes. So vel3 is the velocity in the direction that is 256 long

```python
file = h5.File('dens_data_tcool_0.03.h5', 'r')
dens = file['dens_data'][:]
file.close()
```

```python
# plot 2D slices of the density field
fig, axs = plt.subplots(1, 10, dpi = 400, figsize = (22, 4))
for i in range(10):
    plot = axs[i].imshow(dens[i,-1,:,:,0], cmap = 'Blues', norm=LogNorm(vmin = 0.8, vmax=125), origin='lower')
    axs[i].set_xticks([])
    axs[i].set_yticks([])
    axs[i].set_xlabel('y')
    axs[i].set_title('seed %d' % (i+1))
axs[0].set_ylabel('z')

#add an axis on the right for a colorbar
fig.subplots_adjust(right=0.9)
cbar_ax = fig.add_axes([0.92, 0.15, 0.01, 0.7])
cb = fig.colorbar(plot, cax=cbar_ax)
cb.set_label('density')
```
