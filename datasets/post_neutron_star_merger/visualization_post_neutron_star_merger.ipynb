```python
import h5py
import json
import os
paths = sorted(glob.glob('data/*.hdf5'))
list_of_keys = ['B_unit', 'DTd', 'DTl','DTp','DTr','LEdd','L_unit', 'M_unit', 'Mbh','MdotEdd','RHO_unit','Reh','Rin','Risco','Rout','Rout_rad','Rout_vis','TEMP_unit','T_unit','U_unit','cour','hslope','maxnscatt','mks_smooth','poly_alpha', 'poly_xt','tf']

def read_hdf5_keys_values(file_path):
    """Reads the keys and their values from an HDF5 file."""
    data = {}
    with h5py.File(file_path, 'r') as hdf:
        for key in list_of_keys:
            print(key)
            
            data[key] = float(hdf['metadata']['settings'][key][()])  # Read the value and convert to a Python type
            # else:
            #     data[key] = hdf['metadata']['settings'][key][:]
    return data

def write_to_json(data, output_file):
    """Writes data to a JSON file."""
    with open(output_file, 'w') as json_file:
        json.dump(data, json_file, indent=4)

def process_hdf5_files(file_paths, output_file):
    """Processes multiple HDF5 files and writes their keys and values to a JSON file."""
    all_data = {}
    for file_path in file_paths:
        file_name = os.path.basename(file_path)
        data = read_hdf5_keys_values(file_path)
        all_data[file_name] = data
    write_to_json(all_data, output_file)

# Example usage:
hdf5_files = paths # Replace with your HDF5 file paths
output_json = 'output.json'
process_hdf5_files(hdf5_files, output_json)
```
