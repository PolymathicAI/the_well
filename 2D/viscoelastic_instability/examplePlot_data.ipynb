# Example notebook to read viscoelastic data

Example of EIT data generated using Dedalus v2 by Miguel Beneitez - mb2467@cam.ac.uk <br>
Papers where this data has been used: <br> 
M. Beneitez, J. Page, Y. Dubief & R. R. Kerswell (2024). Multistability of elasto-inertial two-dimensional channel flow. J. Fluid Mech. 981, A30

```python
import h5py                                                                                 
import numpy as np                                                                          
import matplotlib.pyplot as plt   
```

The data includes the variables required ($u,v,p,c_{xx},c_{yy},c_{zz},c_{xy}$) to restart any simulation for a FENE-P fluid in a channel with a constant massflux of $1$.

```python
fname = 'Re1000_Wi50_beta09_Lmax70_Sc500_Lx2pi_Ly2_EIT/file_1.h5' # The file we use
```

Data is stored as Seed x Snapshots x Nx x Ny where Nx and Ny are the number of points in the x and y directions

```python
seed = 1 # Which seed we choose
snapshot = 5 # Which snapshot we choose
```


Note that data should form a time series and the whole series can be reconstructred by reshaping the data so that one seed goes after the previous seed <br>
i.e. 2 x 50 x Nx x Ny corresponds to the 50 snapshots after 1 x 50 x Nx x Ny

Read the data now and show the image (note it's squared now)

```python
file = h5py.File(fname, mode='r')
plt_field = file['c11'][seed,snapshot,:,:] # The data we want to plot
plt.figure()
plt.imshow(plt_field, cmap='viridis')
```
